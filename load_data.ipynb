```python
#Goal: Visualize feature distributions and see patterns for diabetic vs non-diabetic.
import matplotlib.pyplot as plt
import seaborn as sns
#histogram
plt.figure(figsize=(6,4))
sns.histplot(data['Glucose'], kde=True, bins=30)  #kde=True overlays a Kernel Density Estimate — a smooth curve approximating the distribution.
plt.title("Glucose Distribution")
plt.xlabel("Glucose")
plt.ylabel("Count")
plt.show()

# Boxplot of Glucose by Outcome
plt.figure(figsize=(7,4))
sns.boxplot(x='Outcome', y='Glucose', data=data)
plt.title("Glucose by Diabetes Outcome")
plt.show()

'''Interpretation:

People diagnosed with diabetes (Outcome = 1) have significantly higher glucose levels compared to non-diabetics (Outcome = 0).
This confirms that glucose is a strong predictor of diabetes in this dataset.
'''
# Scatterplot BMI vs Age colored by Outcome
plt.figure(figsize=(6,4))
sns.scatterplot(x='Age', y='BMI', hue='Outcome', data=data)
plt.title("BMI vs Age colored by Outcome")
plt.show()

'''Observations:

Diabetic individuals (orange) are more concentrated among older ages and higher BMI values.

Non-diabetic individuals (blue) are spread across younger ages and lower BMI.

The plot shows a positive pattern — as age and BMI increase, the likelihood of diabetes also increases.'''
```

```python
'''Goal:

Compare the average value of a numeric feature between groups.

For diabetes: Compare features (Glucose, BMI, Age, etc.) between Outcome=0 (no diabetes) and Outcome=1 (diabetes).

Since Outcome has 2 groups, we mostly use t-test.
If you had more than 2 groups, you would use ANOVA.'''
from scipy import stats

# Numeric features to test
numeric_features = ['Glucose','BMI','BloodPressure','Insulin','Age']

# Empty list to store results
t_test_results = []

for feature in numeric_features:
    group0 = data[data['Outcome']==0][feature]
    group1 = data[data['Outcome']==1][feature]
    
    # Welch's t-test (does not assume equal variance)
    t_stat, p_val = stats.ttest_ind(group0, group1, equal_var=False)
    t_test_results.append([feature, round(t_stat,3), round(p_val,2)])

# Pretty print the results using tabulate
from tabulate import tabulate
print("\nT-test Results (Welch's) for Numeric Features by Outcome:")
print(tabulate(t_test_results, headers=['Feature','t-statistic','p-value'], tablefmt='fancy_grid'))


'''A Welch’s t-test was performed to identify statistically significant differences in key health indicators between diabetic and non-diabetic patients.
The test revealed that Glucose, BMI, Insulin, and Age were significantly higher among diabetic patients (p < 0.05), suggesting these features are strong predictors of diabetes.'''
```

```python
from scipy import stats
from tabulate import tabulate
# Create Age groups
data['AgeGroup'] = pd.cut(data['Age'], bins=[20,29,39,49,100], labels=['20-28','29-38','39-48','49+'])

# Contingency table: counts of Outcome per AgeGroup
cont_table = pd.crosstab(data['AgeGroup'], data['Outcome'])

#calc chisquare
chi2, p_val, dof, expected = stats.chi2_contingency(cont_table)


print("\nContingency Table: AgeGroup vs Outcome")
print(tabulate(cont_table, headers='keys', tablefmt='fancy_grid'))

print(f"\nChi-square test result: chi2 = {chi2:.3f}, p-value = {p_val:.4f}, dof = {dof}")

#The Chi-square test was significant (χ² = 69.915, p < 0.001, df = 3), indicating that age group is significantly associated with diabetes risk.
```

```python
import pandas as pd
from scipy import stats
from tabulate import tabulate

# ----------- Step 1: Create categorical groups -----------

# Pregnancies group
data['PregnancyGroup'] = pd.cut(data['Pregnancies'], bins=[-1,2,5,20], labels=['0-2','3-5','6+'])

# BMI group (Underweight <18.5, Normal 18.5-24.9, Overweight 25-29.9, Obese >=30)
data['BMIGroup'] = pd.cut(data['BMI'], bins=[0,18.5,24.9,29.9,100], labels=['Underweight','Normal','Overweight','Obese'])

# ----------- Step 2: Create contingency tables and Chi-square tests -----------

def chi_square_test(feature_group, outcome='Outcome'):
    print(f"\nContingency Table: {feature_group} vs {outcome}")
    
    # Create contingency table
    cont_table = pd.crosstab(data[feature_group], data[outcome])
    
    # Pretty print the table
    print(tabulate(cont_table, headers='keys', tablefmt='fancy_grid'))
    
    # Chi-square test
    chi2, p_val, dof, expected = stats.chi2_contingency(cont_table)
    print(f"\nChi-square test result: chi2 = {chi2:.3f}, p-value = {p_val:.4f}, dof = {dof}")
    print("Interpretation: ", "Significant association" if p_val<0.05 else "No significant association")
    
# ----------- Step 3: Run Chi-square tests -----------

# Pregnancies
chi_square_test('PregnancyGroup')

'''Chi-square result: χ² = 40.216, p-value < 0.001, dof = 2
Interpretation:

The p-value is extremely small (<0.05), so we reject the null hypothesis.

This means PregnancyGroup is significantly associated with diabetes.

Observation from table:

Women with 6+ pregnancies have a higher proportion of diabetes (111 diabetics vs 108 non-diabetics).

Women with fewer pregnancies (0-2) have fewer diabetics relative to non-diabetics.

Conclusion: Number of pregnancies is a risk factor for diabetes.
'''
# BMI
chi_square_test('BMIGroup')

'''Chi-square result: χ² = 74.908, p-value < 0.001, dof = 3
Interpretation:

The p-value is extremely small (<0.05), so we reject the null hypothesis.

This indicates a significant association between BMI group and diabetes.

Observations:

Most diabetics are in the Obese group (221 diabetics), followed by Overweight (40).

Normal and Underweight groups have very few diabetics.

Conclusion: Higher BMI strongly increases the risk of diabetes, confirming clinical expectations.'''
```
